## Part 1: Privacy Evaluation

```python
# Metric to measure privacy of selected features (a quasi-identifier would have low privacy / a low AASS)
def average_as_size(df, columns):
    n_users = df.count()[0]
    anonymity_sets = df.groupby(columns).count()
    average_size = n_users / anonymity_sets.count()[0]
    return average_size
```

### Initial Data Analysis

The service provider has access to two datasets, namely _pois.csv_ - the list of Points Of Interest (POIs), some of which (public) can be returned in a response to a query - and _queries.csv_, the history of user queries over the course of 20 days. 

```python
pois_df = pd.read_csv("pois.csv", delim_whitespace=True)

queries_df = pd.read_csv("queries.csv", delim_whitespace=True)
# Add columns for analysis
queries_df["cell_id"] = queries_df.apply(lambda row: location_to_cell_id(row["lat"], row["lon"]), axis=1)
queries_df["day"] = queries_df.apply(lambda row: int(row["timestamp"]) // 24, axis=1)
queries_df["hour"] = queries_df.apply(lambda row: int(row["timestamp"]) - 24*row["day"], axis=1)
queries_df["weekday"] = queries_df.apply(lambda row: row["day"] % 7 < 5, axis=1)
```

### Identifying home and workplace

Since we can now identify queries that originated from a residence or an office, it shouldn't be too difficult to identify a user's home and workplace. We will simply take the home to be the most frequent POI in queries that were issued by the user from an 'appartment_block' or a 'villa'. The approach is similar for the workplace, which is taken to be the most frequent POI in queries issued from an 'office', a 'laboratory' or a 'company'.

```python
home_poi_types = ['appartment_block', 'villa']
work_poi_types = ['office', 'laboratory', 'company']

queries_from_home = queries_with_location[queries_with_location.poi_type.isin(home_poi_types)]
queries_from_work = queries_with_location[queries_with_location.poi_type.isin(work_poi_types)]
```

```python
home_poi_id = queries_from_home.groupby("ip_address")["poi_id"]\
                               .apply(set)\
                               .reset_index()\
                               .rename(columns={"poi_id": "home_id"})

set_sizes = home_poi_id.home_id.apply(lambda s: len(s))
set_sizes[set_sizes != 1]
```

```python
home_poi_id.home_id = home_poi_id.home_id.apply(lambda s: s.pop())
```

```python
work_poi_id = queries_from_work.groupby("ip_address")["poi_id"]\
                               .apply(set)\
                               .reset_index()\
                               .rename(columns={"poi_id": "work_id"})

set_sizes = work_poi_id.work_id.apply(lambda s: len(s))
set_sizes[set_sizes != 1]
```

In turns out that our simulated users are quite simple, they only visited a single POI that would be considered a residence, and a single POI that would be considered a workplace. We can now map every IP address to a (home_id, work_id) tuple.

```python
work_poi_id.work_id = work_poi_id.work_id.apply(lambda s: s.pop())
```

```python
user_meta = home_poi_id.merge(work_poi_id, on="ip_address")
user_meta
```

```python
duplicates = user_meta[user_meta.duplicated(["home_id", "work_id"], keep=False)].sort_values("home_id").reset_index()
print(duplicates.count()[0], "duplicates")
```

All further findings can be traced back to identified users. We can even make it a unique identifier by adding more top visited POIs. 

### Hobbies

Once we have a quasi-identifier, we can focus on extracting sensitive information from the user's spatio-temporal history of queries. To find out about a user's hobbies, we look at the types of POIs visited or queried. 

```python
full_queries = queries_with_location.merge(user_meta, on="ip_address")
full_queries.head()
```

```python
queries_from_misc = full_queries[(full_queries.poi_id != full_queries.home_id) & (full_queries.poi_id != full_queries.work_id)]
queries_from_misc.head()
```

```python
visits = queries_from_misc.groupby(["ip_address", "poi_type"]).size().unstack('poi_type', fill_value=0)
visits["total"] = visits.sum(axis=1)
visits = visits.sort_values(by="total", ascending=False)
visits.head()
```

```python
query_pois = queries_from_misc.groupby(["ip_address", "poi_type_query"]).size().unstack('poi_type_query', fill_value=0)
query_pois["total"] = query_pois.sum(axis=1)
query_pois = query_pois.sort_values(by="total", ascending=False)
query_pois.head()
```

We can see that some users have what would be considered an unhealthy lifestyle, with frequent visits of bars and clubs over the course of 20 days. 

We would now like to find the two most popular POIs, that are not the presumed home or workplace.

```python
user_poi_queries = queries_from_misc.groupby(['ip_address','poi_id']).size().reset_index(name="count")
user_poi_queries["count_id"] = user_poi_queries.apply(lambda row: (row["count"], row["poi_id"]), axis=1)
user_poi_queries.head()
```

```python
top_pois = user_poi_queries.groupby("ip_address")["count_id"]\
                           .apply(lambda s: list(map(lambda x: x[1], sorted(list(s), reverse=True)[:2])))\
                           .reset_index(name="top_cells")
top_pois[['misc1_id', 'misc2_id']] = pd.DataFrame(top_pois['top_cells'].tolist(), index=top_pois.index)
top_pois = top_pois.drop(columns=["top_cells"])
```

```python
user_meta = user_meta.merge(top_pois, on="ip_address")
user_meta.head()
```

```python
user_meta[user_meta.duplicated(["home_id", "work_id", "misc1_id"], keep=False)]
```

```python
average_as_size(user_meta, ["home_id", "work_id", "misc1_id", "misc2_id"])
```

As mentioned, adding information about the top visited POIs returns a unique identifier that can be used to reveal the user's identity. 

### Info about size of company, how many people are working there

Once we have the alleged POI IDs of a user's home and workplace, we can tell how many people are living in a household or working for a company. 

```python
residents = user_meta.groupby("home_id")["ip_address"]\
                     .apply(list)\
                     .reset_index()

offices = user_meta.groupby("work_id")["ip_address"]\
                   .apply(list)\
                   .reset_index()\
                   .rename(columns={"ip_address": "employees"})

offices["size"] = offices.employees.apply(lambda x: len(x))
offices = offices.sort_values(by="size", ascending=False)
offices.head()
```

### Partners and colleagues 

Another thing we can do is find out who a user met with, by looking for queries from the same POI, shortly after or before the user sent a query.

```python
df1 = queries_with_location[["ip_address", "lat", "lon", "timestamp"]]
df2 = df1.copy()

same_loc = df1.merge(df2, on=["lat", "lon"])
same_loc = same_loc[same_loc.ip_address_x != same_loc.ip_address_y]
same_loc["diff"] = (same_loc.timestamp_x - same_loc.timestamp_y).abs()
same_loc = same_loc[same_loc["diff"] < 0.5]
```

```python
same_loc
```

```python
closeness_counts = same_loc.groupby(["ip_address_x", "ip_address_y"])[["lat"]].count().rename(columns={"lat": "closeness_count"})
closeness_counts.head()
```

If the count of queries from the same place in a small time window is high, that the two users are likely not there by coincidence but know each other.

## Part 2: Defence

The proposed client-side defence is to send cell IDs (instead of exact locations) to the service provider, and increase the size of the cells. We would like to find out how this affects privacy in terms of AASS, based on presumed cell IDs of home and workplace (and other top visited cells). Home and workplace cell IDs can be found using common sense, the cell of workplace is most likely the cell where a user spents most of his time during working hours (weekdays, 9am - 5pm).

```python
# Modified version of the function provided in grid.py, that takes as argument the number of cells
def location_to_cell_id(lat, lon, cell_num_lat, cell_num_lon):
    grid_lat_points = [MAP_LAT + i * (MAP_SIZE_LAT / cell_num_lat)
                       for i in range(1, cell_num_lat + 1)]
    grid_lon_points = [MAP_LON + i * (MAP_SIZE_LON / cell_num_lon)
                       for i in range(1, cell_num_lon + 1)]
    
    """Get the grid cell ID for a given latitude and longitude."""
    if not (MAP_LAT <= lat < MAP_LAT + MAP_SIZE_LAT) or not (
        MAP_LON <= lon < MAP_LON + MAP_SIZE_LON
    ):
        raise ValueError("Out of area range.")

    i = bisect.bisect(grid_lat_points, lat)
    j = bisect.bisect(grid_lon_points, lon)
    return i * cell_num_lat + j + 1

# Returns two cell IDs that are not those of the presumed home and workplace, -1 if none exist
def filter_top_cells(home_cell_id, work_cell_id, top_cells):
    filtered = [cell_id for cell_id in top_cells if not (cell_id == home_cell_id or cell_id == work_cell_id)]
    missing_ids = 2 - len(filtered)
    if missing_ids > 0:
        for i in range(missing_ids):
            filtered.append(-1)
    return filtered[:2]
```

```python
pois_mod_df = pd.read_csv("pois.csv", delim_whitespace=True)
queries_mod_df = pd.read_csv("queries.csv", delim_whitespace=True)
queries_mod_df["day"] = queries_mod_df.apply(lambda row: int(row["timestamp"]) // 24, axis=1)
queries_mod_df["hour"] = queries_mod_df.apply(lambda row: int(row["timestamp"]) - 24*row["day"], axis=1)
queries_mod_df["weekday"] = queries_mod_df.apply(lambda row: row["day"] % 7 < 5, axis=1)
n_queries = queries_mod_df.count()[0]
```

```python
home_work_aass = []
misc1_aass = []
misc2_aass = []
avg_response_number = []

cell_numbers = list(range(2, 11))

for n_cells in cell_numbers:
    print("Number of cells:", n_cells, "x", n_cells)
    pois_mod_df["cell_id"] = pois_mod_df.apply(lambda row: location_to_cell_id(row["lat"], row["lon"], n_cells, n_cells), axis=1)
    queries_mod_df["cell_id"] = queries_mod_df.apply(lambda row: location_to_cell_id(row["lat"], row["lon"], n_cells, n_cells), axis=1)
    
    # Working hours: 9 to 5 job, on weekdays
    working_hours = (queries_mod_df.weekday) & (8 < queries_mod_df.hour) & (queries_mod_df.hour < 17)
    working_queries = queries_mod_df[working_hours]
    leisure_queries = queries_mod_df[~working_hours]
    
    # Cell ID of the presumed home is the most frequent cell ID during leisure time
    home_mod_poi_id = leisure_queries.groupby(['ip_address','cell_id']).size().groupby(level=0).idxmax()\
                                .apply(lambda x: x[1]).reset_index(name='home_cell_id')
    # Cell ID of the presumed workplace is the most frequent cell ID during working hours
    work_mod_poi_id = working_queries.groupby(['ip_address','cell_id']).size().groupby(level=0).idxmax()\
                                .apply(lambda x: x[1]).reset_index(name='work_cell_id')
    
    user_meta_mod = home_mod_poi_id.merge(work_mod_poi_id, on="ip_address")
    # Add info about a user's presumed home and workplace to the queries
    full_queries_mod = queries_mod_df.merge(user_meta_mod, on="ip_address")
    user_cell_queries = full_queries_mod.groupby(['ip_address','cell_id', 'home_cell_id', 'work_cell_id']).size().reset_index(name="count")
    user_cell_queries["count_id"] = user_cell_queries.apply(lambda row: (row["count"], row["cell_id"]), axis=1)
    
    # Find top cells
    top_cells = user_cell_queries.groupby(['ip_address', 'home_cell_id', 'work_cell_id'])["count_id"]\
                                 .apply(lambda s: list(map(lambda x: x[1], sorted(list(s), reverse=True))))\
                                 .reset_index(name="top_cells")
    # Keep only cells that are not presumed cells of home and workplace, and fill missing values
    top_cells["top_cells"] = top_cells.apply(lambda row: filter_top_cells(row["home_cell_id"], row["work_cell_id"], row["top_cells"]), axis=1)
    top_cells[['misc1_cell_id', 'misc2_cell_id']] = pd.DataFrame(top_cells['top_cells'].tolist(), index=top_cells.index)
    top_cells = top_cells.drop(columns=["top_cells"])
    
    # Compute metrics we are interested in
    home_work_aass.append(average_as_size(top_cells, ["home_cell_id", "work_cell_id"]))
    misc1_aass.append(average_as_size(top_cells, ["home_cell_id", "work_cell_id", "misc1_cell_id"]))
    misc2_aass.append(average_as_size(top_cells, ["home_cell_id", "work_cell_id", "misc1_cell_id", "misc2_cell_id"]))
    
    responses = queries_mod_df.merge(pois_mod_df, left_on=["cell_id", "poi_type_query"], right_on=["cell_id", "poi_type"]).count()[0]
    avg_response_number.append(responses / n_queries)
```

```python
data = (home_work_aass, misc1_aass, misc2_aass)
colors = ("red", "green", "blue")
groups = ("Home and workplace", "With first most visited place", "With second most visited place")

# Create plot
fig = plt.figure(figsize=(12,7))
ax = fig.add_subplot(1, 1, 1)

x = [n_cells**2 for n_cells in cell_numbers]

for y, color, group in zip(data, colors, groups):
    ax.scatter(x, y, alpha=0.8, c=color, edgecolors='none', label=group)

plt.xlabel('Number of grid cells')
plt.ylabel('AASS')
plt.legend(loc=1)
#plt.show()
fig.savefig('aass_scatter.png')
```

```python
fig = plt.figure(figsize=(12,7))
ax = fig.add_subplot(1, 1, 1)
ax.scatter(x, avg_response_number, c=(0,0,0), alpha=0.8)
plt.xlabel('Number of grid cells')
plt.ylabel('Avg. number of responses')
#plt.show()
fig.savefig('responses_scatter.png')
```
